```python
from fastbook import *
from fastai.vision.widgets import *
from duckduckgo_search import ddg_images
from fastcore.all import *

def search_images(term, max_images=30):
    print(f"Searching for '{term}'")
    return L(ddg_images(term, max_results=max_images)).itemgot('image')
```

## Import tree dictionary

```python
import pickle 

file = open("dict_tree.pkl",'rb')
dict_tree = pickle.load(file)

latin_names = list(dict_tree.keys())
common_names = list(dict_tree.values())
```

```python
dict_tree[latin_names[0]]
```

## Search images

```python
name = common_names[3]
```

```python
#NB: `search_images` depends on duckduckgo.com, which doesn't always return correct responses.
#    If you get a JSON error, just try running it again (it may take a couple of tries).
urls = search_images(name+" leaf", max_images=150)
urls[1]
```

```python
dest = 'images/leaf.jpg'
download_url(urls[3], dest, show_progress=False)

im = Image.open(dest)
im.to_thumb(256,256)
```

## Download all images

```python
tree_types = latin_names
path = Path('tree_images')
```

```python
from time import sleep
from tqdm import tqdm

if not path.exists():
    path.mkdir()
    for tree in tqdm(tree_types):
        print(f'Downloading {tree}: {dict_tree[tree]}')
        dest = (path/tree)
        dest.mkdir(exist_ok=True)
        url_list = search_images(f'{dict_tree[tree]} leaf', max_images=30)
        print(str(len(url_list))+" urls found")
        download_images(dest, urls=url_list)
        sleep(2.)
```

## Turn all images into jpg

```python
from PIL import Image 
import os 

path = 'tree_images'
```

```python
for folder in os.listdir(path):
    try:
        file_list=os.listdir(f'{path}/{folder}')

        for file in file_list: 
            if file.endswith(".png"): 
                file_path = f'{path}/{folder}/{file}'
                file_name, file_ext = os.path.splitext(file_path)
                im = Image.open(file_path)
                # Convert png to jpg
                rgb_im = im.convert('RGB')
                rgb_im.save(f'{file_name}.jpg')

                # Remove png
                os.remove(f'{file_name}.png')
    except:
        None
```
